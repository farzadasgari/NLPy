# Part-of-Speech (PoS) Tagging

Part-of-Speech (PoS) tagging is the process of assigning a part of speech to each word in a sentence. PoS tags include categories such as noun, verb, adjective, adverb, etc.

## Parts of Speech

There are **eight** main parts of speech:

### 1. Nouns
**Definition**: Nouns are words that name people, places, things, or ideas.
**Examples**: 
- **People**: teacher, student, doctor
- **Places**: school, city, park
- **Things**: book, computer, apple
- **Ideas**: freedom, love, happiness

**Sentence Example**:
"The **teacher** gave the **students** new **books** to read in the **classroom**."

### 2. Pronouns
**Definition**: Pronouns are words that replace nouns to avoid repetition.
**Examples**: he, she, it, they, him, her, them

**Sentence Example**:
"**She** gave **him** a gift because **he** helped **her** with **her** project."

### 3. Adjectives
**Definition**: Adjectives are words that describe or modify nouns.
**Examples**: happy, blue, tall, quick

**Sentence Example**:
"The **happy** child played with the **blue** ball on the **tall** slide."

### 4. Verbs
**Definition**: Verbs are words that express actions, occurrences, or states of being.
**Examples**: run, jump, think, is, are

**Sentence Example**:
"The children **run** and **jump** in the park while their parents **watch**."

### 5. Adverbs
**Definition**: Adverbs are words that modify verbs, adjectives, or other adverbs, often indicating how, when, where, or to what extent.
**Examples**: quickly, very, well, yesterday

**Sentence Example**:
"She **quickly** finished her homework, so she could play outside **happily**."

### 6. Prepositions
**Definition**: Prepositions are words that show relationships between nouns (or pronouns) and other words in a sentence, indicating direction, place, time, etc.
**Examples**: in, on, at, by, for, with

**Sentence Example**:
"The book is **on** the table, and the cat is **under** the chair."

### 7. Conjunctions
**Definition**: Conjunctions are words that connect words, phrases, or clauses.
**Examples**: and, but, or, because, although

**Sentence Example**:
"I wanted to go for a walk, **but** it started raining."

### 8. Interjections
**Definition**: Interjections are words or phrases that express strong emotion or surprise.
**Examples**: oh, wow, ouch, hey

**Sentence Example**:
"**Wow!** That was an amazing performance."

---

### Summary of Sentences

To summarize these parts of speech within a single context:

**Complete Example**:
"**Wow!** The **happy** **teacher** quickly gave **his** **students** new **books** in the **classroom**, and **they** were excited because **they** love to read."

- **Nouns**: teacher, students, books, classroom
- **Pronouns**: his, they
- **Adjectives**: happy, new
- **Verbs**: gave, love
- **Adverbs**: quickly
- **Prepositions**: in
- **Conjunctions**: and, because
- **Interjections**: Wow!

## Why PoS Tagging?

PoS tagging helps in:
- Understanding the structure and meaning of a sentence.
- Enhancing text preprocessing for other NLP tasks.
- Improving the performance of text-based models by providing syntactic information.

### Explanation of `token.pos_`

- **Coarse-Grained PoS Tag**: The `pos_` attribute represents the general category of the word's function in the sentence, such as noun, verb, adjective, etc.
- **String Format**: The tag is returned as a string, which is easier to interpret compared to the integer code (`token.pos`).

#### Common Coarse-Grained PoS Tags

- **ADJ**: Adjective
- **ADP**: Adposition (prepositions and postpositions)
- **ADV**: Adverb
- **AUX**: Auxiliary verb
- **CONJ**: Coordinating conjunction
- **DET**: Determiner
- **INTJ**: Interjection
- **NOUN**: Noun
- **PROPN**: Proper noun
- **VERB**: Verb
- **PRON**: Pronoun

#### `token.pos_` vs. `token.tag_`

- **`token.pos_`**: Provides a coarse-grained, general category of the word's PoS.
- **`token.tag_`**: Provides a fine-grained PoS tag that includes more specific details about the word's role in the sentence.

For most NLP tasks, `token.pos_` is sufficient, but `token.tag_` is used when more detail is needed, such as distinguishing between different forms of verbs or pronouns.
